```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
pd.set_option("display.max_columns", 1000)
```

```python
df = pd.read_csv("C:\\Users\\Mubasshira\\Downloads\\Hr_analytics.csv")
df.head()
```

```python
df.drop(columns=['EmployeeCount', 'Over18'], inplace=True)
```

```python
def learnings(col):
    if col == 1:
        return 'Below College'
    elif col == 2:
        return 'College'
    elif col == 3:
        return 'Bachelor'
    elif col == 4:
        return 'Master'
    elif col == 5:
        return 'Doctor'
    
df["education_level"] = df['Education'].apply(learnings)
```

```python
df
```

```python
def envsat(col):
    if col == 1:
        return 'Low'
    elif col == 2:
        return 'Medium'
    elif col == 3:
        return 'High'
    elif col == 4:
        return 'Very High'
    
df['environmental_satiscation'] = df['EnvironmentSatisfaction'].apply(envsat)
```

```python
job_map = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
performance_map = {1: 'Low', 2: 'Good', 3: 'Excellent', 4: 'Outstanding'}
wlb_map = {1: 'Bad', 2: 'Good', 3: 'Better', 4: 'Best'}

df['job_involvement_label'] = df['JobInvolvement'].map(job_map)
df['job_satisfaction_label'] = df['JobSatisfaction'].map(job_map)
df['performance_rating_label'] = df['PerformanceRating'].map(performance_map)
df['relationship_satisfaction_label'] = df['RelationshipSatisfaction'].map(job_map)
df['work_life_balance_label'] = df['WorkLifeBalance'].map(wlb_map)
```

```python
df.drop(columns=['Education','EnvironmentSatisfaction', 'JobInvolvement', 'JobSatisfaction', 'PerformanceRating',
                 'RelationshipSatisfaction', 'WorkLifeBalance'], inplace=True)
```
